```python
import cvxpy as cp
import numpy as np


# Construct the problem.
x = cp.Variable(1)
y = cp.Variable(1)
objective = cp.Minimize(cp.square(x-1.5) + cp.square(y-2.5))
constraints = [0 <= x, x <= 2, 1 <= y, y<= 2, x**2 + y**2 <= 4]
prob = cp.Problem(objective, constraints)

# The optimal objective value is returned by `prob.solve()`.
result = prob.solve()
# The optimal value for x is stored in `x.value`.
print(x.value)
```
